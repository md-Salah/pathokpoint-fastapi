```python
import pandas as pd
```

```python
df = pd.read_csv('seba.csv')
```

```python
df.dropna(subset=['images'], inplace=True)
```

```python
base_url = 'http://mdsalah.customerserver003003.eurhosting.net/WATERMARKED-2/'
```

```python
df['images'] = df['images'].apply(lambda x:  base_url + x.strip().replace(' ', '%20'))
```

```python
df.head(1)
```

```python
for col in ['sku', 'name', 'edition', 'notes', 'cover', 'ln', 'condition']:
    df[col] = df[col].astype(str)
```

```python
df.dtypes
```

```python
df.fillna('', inplace=True)
df.replace('nan', '', inplace=True)
```

```python
df['sale_price']
```

```python
df.drop(columns=['id', 'source_url', 'name_alt'])
```

```python
df.to_csv('seba_clean.csv', index=False, encoding='utf-8-sig')
```
